```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Start with one review:

authors = ['lucas','kiyan', 'arvind', 'sam', 'madridista']
texts = {}
for author in authors:
    with open('mm_text_'+author+'.txt', 'r', encoding='utf-8') as file:
        texts[author] = file.read()
```

```python
for author in authors:
    # Create and generate a word cloud image:
    wordcloud = WordCloud(max_font_size=40, max_words=100, background_color="black").generate(texts[author])
    # Display the generated image:
    plt.figure(figsize=(15,15))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(author)
    plt.show()
```
